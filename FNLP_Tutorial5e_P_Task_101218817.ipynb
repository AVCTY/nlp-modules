##**Discussions**:
<p align=justify>Word embeddings are a very popular way to represent document vocabulary. This 
is because it has the capability to capture the context of a waord in a document, semantic and syntactic similarity, relation with other words. Loosely speaking, word embeddings are vector representations of a particular word. One of the popular methods to learn word embeddings is Word2Vec whereby it uses a shallow neural network which was developed by Tomas Mikolov in 2013 at Google.There are also other methods of word embeddings out there as well such as GloVe [1].</p>

<p align=justify>For Word2Vec, it uses a small neural network to calculate word embeddings based on the context of the words and there two approches to implement this. The first being the continuous bag of words (CBOW) where this approach, the network tries to predict which word is most likely given the context and words that are likely to appear can be interpreted as having a shared dimension. The second approach is skip-gram whereby the network tries to use the target word to predict its context [2].</p>

<p align=justify>While Word2Vec learns embeddings by relating target words to their context and a frequent co-occurence of words creates more training examples, it does not carry additional information. In contrast to this, GloVe emphasizes that the frequency of co-occurences is vital information and should not be wasted as additional training examples. GloVe builds word embeddings by combining word vectors which relates directly to the probability of these word's co-occurence in the corpus [2].</p>
